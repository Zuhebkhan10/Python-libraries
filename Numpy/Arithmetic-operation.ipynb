```python
# Arithmetic operations are used for numerical computation and we can perform them on arrays using NumPy

# Addition of Arrays
import numpy as np
num1=np.array([10,20,30])
num2=np.array([50,60,70])

res=np.add(num1,num2)
print("Sum of array is ",res)
```

```python
# Subtraction of Arrays
import numpy as np
num1=np.array([110,120,130])
num2=np.array([50,60,70])

res=np.subtract(num1,num2)
print("Sum of array is ",res)
```

```python
# Multiplication of Arrays
import numpy as np
num1=np.array([10,2,3])
num2=np.array([50,60,70])

res=np.multiply(num1,num2)
print("Sum of array is ",res)
```

```python
# Division of Arrays
import numpy as np 
num1=np.array([10,20,40])
num2=np.array([5])

res=np.divide(num1,num2)
print(res)

print(res.dtype)
```

```python
# Exponentiation (Power)
import numpy as np 
num1=np.array([2,5,11])
num2=np.array([2,3,3])

res=np.power(num1,num2)
print("Power is ",res)
```

```python
# Modulus Operation:(Reminders)
import numpy as np 
num1=np.array([2,15,110])
num2=np.array([2,4,20])

res=np.mod(num1,num2)
print("Reminder is ",res)
```
