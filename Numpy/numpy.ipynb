```python
"""when we add a single element array to the existing array of different size then the new array element 
is going add up with each element in exisiting array this method is known as "Broadcasting method
"""

#Normal way broadcasting-method
square=[1,2,3,4,5,6]
#print above list with product of 2
for i in square:
    print(i*2,end=" ")


# way to broadcasting-method 
bc=np.array(square)
print(bc*2)
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])
#combine the above 2 arrays
arr3 = np.append(arr1,arr2)
print(arr3)


#find the sqrt of arr1
print(np.sqrt(arr1))
```

```python
# Dot product
import numpy as np
a=np.array([1,2])
b=np.array([3,4])
dot=0

for i in range(len(a)):
    dot+=a[i]*b[i]
print(dot)    
```

```python
#define dot product?
#vector multiplication of array's and sum  of the result
#multiply each corresponding element and do the sum

# easy with numpy
dot=np.dot(a,b)
print(dot)

#manual method
c=a*b
print(c)

d=np.sum(c)
print(d)

e=a.dot(b)
print(e)

#create an array with set
s={20,30,40,50}
print(np.array(s))
```
