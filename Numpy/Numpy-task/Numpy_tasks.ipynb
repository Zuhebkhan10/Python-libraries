#### 27. Consider an integer vector Z, which of these expressions are legal? (★☆☆)

```python
Z**Z
2 << Z >> 2
Z <- Z
1j*Z
Z/1/1
Z<Z>Z
```

````python
import numpy as np
Z = np.array([1, 2, 3, 4, 5])

res = Z ** Z
print("Z**Z =", res)


shift_res=2<<Z>>2
print("2<<Z>>2=",shift_res)

comparison_res = Z < -Z
print("Z < -Z =", comparison_res)


complex_res = 1j * Z
print("1j * Z =", complex_res)


division_res = Z / 1 / 1
print("Z / 1 / 1 =", division_res)
````

#### 33. How to get the dates of yesterday, today and tomorrow? (★☆☆)
(**hint**: np.datetime64, np.timedelta64)

````python
import numpy as np

today=np.datetime64('today')
yesterday=today-np.timedelta64(1,'D')
tomorrow=today+np.timedelta64(1,'D')

print("Today date is ",today)
print("yesterday date was",yesterday)
print("Tomorrow date will be ",tomorrow)
````

#### 35. How to compute ((A+B)\*(-A/2)) in place (without copy)? (★★☆)
(**hint**: np.add(out=), np.negative(out=), np.multiply(out=), np.divide(out=))

````python
A=np.array([12.,13.,14.,15.])
B=np.array([6.,7.,9.,10.])

np.add(A,B,out=A)

np.negative(B,out=B)

# np.multiply(A,B,out=A)

np.divide(A,2, out=B)
````

#### 36. Extract the integer part of a random array using 5 different methods (★★☆)
(**hint**: %, np.floor, np.ceil, astype, np.trunc)

````python
n=np.random.uniform([-10,20,5])
print(n)


p=n-(n%2)
print("Random percentage is",p)

f=np.floor(n)
print("random floor value is ",f)

c=np.ceil(n)
print("random ceil value is ",c)

s=n.astype(int)
print(s)

t=np.trunc(n)
print(t)


````
