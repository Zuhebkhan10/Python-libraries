```python
 # Broadcasting a Scalar to a 1D Array
import numpy as np
arr=np.array([1,2,4,5])
res=arr+1
print(res)
```

```python
# Broadcasting a 1D Array to a 2D Array
import numpy as np
a=np.array([2,4,6])
b=np.array([[5,10,15],[20,30,40]])
res=a+b
print(res)
```

```python
# Broadcasting in Conditional Operations
import numpy as np
a=np.array([18,21,13,23,15])
b=np.array(["Adult","Minor"])

res=np.where(a>=18,b[0],b[1])
print(res)
```

```python
# Using Broadcasting for Matrix Multiplication
import numpy as np
a=np.array([2,4,6])
b=np.array([5,10,15])
res=a*b
print(res)
```

```python
 # Scaling Data with Broadcasting
import numpy as np
fd = np.array([[0.2, 2.5, 3.9],
                [52.4, 23.6, 36.5],
                [55.2, 31.7, 23.9],
                [14.4, 11.0, 4.9]])
cpg=np.array([2,3,4])

res=fd*cpg
print("Multiplication is \n",res)
```

```python
 # Normalizing Image Data
import numpy as np
img = np.array([ [1000, 120, 130],
                 [900, 110, 140],
                 [80, 100, 120]])

m=img.mean(axis=0)
n=img.std(axis=1)

res=(img-m)/n
print(res)

               
```
