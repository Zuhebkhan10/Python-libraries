```python
# Dot-product: dot product 

import numpy as np
a=np.array([[1,2],[1,4]])
b=np.array([[3,6],[2,5]])
#Apply the dot product on above array's

c=np.dot(a,b)
print(c)
d=a@b
# print(d)
```

```python
# Diagonal
# Print find the digonal element in array is
d=np.diag(a)
print(d)
```

```python
# Boolean 
# Check the elements of aray "a" greater then 2
a=np.array([[1,2,3],
            [6,4,5]]) # Array direct apply the conditional statement by numpy-method
print(a>2)


# print the numbers greater than 2
print(a[a>2])
```

```python
# check the number greater than "4"
l=[2,4,5,7,9]# List apply the using for loop

for i in l:
    if i>4:
        print(True)
    else:
        print(False)
    
```
