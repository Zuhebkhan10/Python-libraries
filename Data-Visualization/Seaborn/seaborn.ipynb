```python
# Seaborn :
# Seaborn is a pyhton library built an function of matplotlib that make if easier  and prettier to create 
# complex beautiful visualization.

# what do you need
# less code
# better looking graps
# easy handling of dataframe like from pandas

# It is mainly used for
# Data analysis (EDA – Exploratory Data Analysis)
# Statistical plotting
# Pattern & relationship detection
# Distribution visualization



# types of Seaborn
# Distribution Plots
# Used to understand data spread (single variable)

# histplot()
# Use: Frequency of values
# Example: Salary distribution, delivery time

# kdeplot()
# Use: Smooth distribution curve
# Example: Probability density

# displot()
# Use: Figure-level distribution (multiple histograms)

# rugplot()
# Use: Show individual data points on axis


# Relational Plots:
# Used to find relationship between numeric variables

# scatterplot()
# Use: Correlation between two variables
# Example: Height vs Weight, Ad spend vs Sales

# lineplot()
# Use: Trend over time
# Example: Monthly revenue, stock price
```

```python
# . Categorical Plots
# Used to compare categories

# barplot()
# Use: Average value per category
# Example: Avg salary by department

# countplot()
# Use: Count of records per category
# Example: Orders per city

# boxplot()
# Use: Median, quartiles, outliers
# Example: Salary outlier detection

# violinplot()
# Use: Distribution + boxplot together

# stripplot()
# Use: Scatter for categorical data

# swarmplot()
# Use: Non-overlapping categorical scatter

# 4. Matrix Plots
#  Used for correlation & missing values
# heatmap()
# Use: Correlation matrix, missing value visualization

# clustermap()
# Use: Group similar rows/columns

# Regression Plots
# regplot()

# Use: Scatter + regression line
# Shows correlation strength

# lmplot()
# Use: Regression with categories
```
