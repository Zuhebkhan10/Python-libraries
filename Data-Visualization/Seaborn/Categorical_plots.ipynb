```python
# barplot()
# Use: Average value per category
import matplotlib.pyplot as plt
import seaborn as sns

df=sns.load_dataset("tips")
df.head()
# data.shape
# data.info()
data.describe()

sns.barplot(data=df,x='sex',y='total_bill',hue='time')
plt.show()
df.head()
```

```python
# countplot()
# Use: Count of records per category


sns.countplot(data=df,x='total_bill',hue='time')
plt.show()
```

```python
# boxplot()
# Use: Median, quartiles, outliers
# Example: Salary outlier detection
df=sns.load_dataset("tips")
df.head()
# data.shape
# data.info()
data.describe()

sns.boxplot(data=df,x='sex',y='total_bill',hue='day')
plt.show()
df.head()
```

```python
# violinplot()
# Use: Distribution + boxplot together

sns.violinplot(data=df,x='sex',y='total_bill',hue='day')
plt.show()
```

```python
sns.scatterplot(data=df,x='tip',y='total_bill',hue='sex',style='time',size='size')
plt.show()
```

```python
# stripplot()
# Use: Scatter for categorical data
sns.stripplot(data=df,x='sex',y='total_bill',hue='time')
plt.show()
```

```python
# swarmplot()
# Use: Non-overlapping categorical 
sns.swarmplot(data=df,x='sex',y='tip',hue='time')
plt.show()
```
