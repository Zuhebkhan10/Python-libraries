```python
# histplot()
# Use: Frequency of values

import matplotlib.pyplot as plt
import seaborn as sns

df=sns.load_dataset('tips')
# df.head()

sns.histplot(data=df,x='total_bill',hue='sex',kde=True)
plt.title("Total bill by gender")
plt.show()
df.head()
```

```python
# displot()
# Use: Figure-level distribution (multiple histograms)

sns.displot(data=df,x='total_bill',hue='time',kde=True)
plt.show()
```

```python
# rugplot()
# Use: Show individual data points on axis
sns.rugplot(data=df,x='total_bill')
plt.show()
```

```python
# kdeplot()
# Use: Smooth distribution curve
# Example: Probability density
sns.kdeplot(data=df,x='total_bill')
plt.show()
```
