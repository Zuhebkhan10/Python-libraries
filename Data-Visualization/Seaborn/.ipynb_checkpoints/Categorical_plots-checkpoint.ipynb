```python
# boxplot()
# Use: Median, quartiles, outliers
# Example: Salary outlier detection
df=sns.load_dataset("tips")
df.head()
# data.shape
# data.info()
data.describe()

sns.boxplot(data=df,x='sex',y='total_bill',hue='day')
plt.show()
df.head()
```
