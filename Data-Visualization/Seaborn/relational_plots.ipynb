```python
# relational plots :used to find relationship between numeric variables

# scatterplot()
# Use: Correlation between two variables

import matplotlib.pyplot as plt
import seaborn as sns


df=sns.load_dataset('flights')
df.head()
# df.info()
# df.describe()
sns.scatterplot(data=df,x='year',y='passengers',hue='month')
plt.show()
```
