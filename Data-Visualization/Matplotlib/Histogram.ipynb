```python
# Histogram: A histogram is a type of plot that show the distribution of a datset.
# Its especially useful for visualization the frequrency of numerial data within specified ranges

# Data spread
# Skewness
# Outliers
# Most frequent range
```

```python
import matplotlib.pyplot as plt

salaries = [25000, 30000, 40000, 45000, 48000, 50000,
            52000, 60000, 75000, 90000, 120000]

plt.hist(salaries, bins=10, edgecolor="black")

plt.xlabel("Salary Range")
plt.ylabel("Number of Employees")
plt.title("Salary Distribution Histogram")

plt.show()

# Interview-ready explanation:
# Histograms are used to analyze the distribution of continuous data, such as salaries, customer ages,
# or delivery times, to identify patterns, skewness, and the most frequent value ranges.
```

```python
import matplotlib.pyplot as plt

delivery_time = [18, 22, 25, 27, 29, 30, 32, 35, 40, 45, 20, 24, 28, 31, 33]

plt.hist(delivery_time, bins=5, edgecolor="black")

plt.axvline(x=20, color="red", linestyle="--", label="SLA Limit (30 min)")

plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Number of Orders")
plt.title("Delivery Time Distribution")
plt.legend()

plt.show()

# bins--> Number or custom edges of bins
# edgecolor--> color around each bar
# axvline--> vertical reference

# Interview-ready explanation:
# We used a histogram to analyze delivery time distribution and added an SLA reference line at 30 minutes.
# Most deliveries fall within the SLA, but a small percentage exceed it, indicating opportunities for route
# optimization and resource planning.
```
