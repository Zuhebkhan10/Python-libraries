```python
import matplotlib.pyplot as plt

years=[2014,2016,2018,2019,2020,2022,2023,2025,2026]
runs=[900,320,550,469,690,760,890,1010,1120]

plt.bar(years,runs)
plt.xlabel('years')
plt.ylabel('runs')
plt.title('Virat Kohli yearly runs')
plt.show()
```

```python
# Setting bar width $ side by side bar charts
import matplotlib.pyplot as plt
import numpy as np
Virat=[500,600,620,740,660,889,878,1020,1240]
Rohit=[430,490,560,660,780,910,1040,1122,1230]
Hardik=[230,340,405,565,430,560,890,250,939]

res=np.arange(len(years)) # index position
width=0.25
# print(res)

# Plotting side by side bar chart
plt.bar(res-width,Virat,width=width,label="Virat")
plt.bar(res+width,Rohit,width=width,label="Rohit")
plt.bar(res+width,Hardik,width=width,label="Hardik")
plt.xlabel('years')
plt.ylabel('Runs')
plt.title('Runs Comparison')
plt.xticks(res,years) #uses numeric res-values and show actual year instead of 
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# A retail company wants to analyze Online vs offline Store sales for the first quarter (Jan–Mar) to:
# Identify the better performing channel
# Check monthly growth
# Support decision making for marketing budge

import matplotlib.pyplot as plt
import numpy as np

months = ["Jan", "Feb", "Mar"]
online = [50000, 60000, 55000]
offline = [40000, 45000, 50000]

x = np.arange(len(months))
width = 0.25

plt.bar(x - width/2, online, width, label="Online Stores Sales")
plt.bar(x + width/2, offline , width, label=" Offline Store Sales")

plt.xticks(x, months)
plt.xlabel('Months')
plt.ylabel("Sales")
plt.title("Online vs Ofline Store Sales Comparison")
plt.legend()

plt.show()
```
