```python
#load the inbuilt data
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
# df.tail()
# df.info()
# df.describe()
```

```python
#Draw a line plot between tips and total bill?
#plt.plot(x,y,'o:r')
plt.plot(df['tip'],df['total_bill'],'o:r')
plt.show()
#when we want to see the distribution of 2 continuous variables(columns) we use scatter plot
```

```python
# Scatter plot
plt.scatter(df['tip'],df['total_bill'],s=45,color="r")
plt.show()
```

```python
# sns.scatter plot (data=)
sns.scatterplot(data=df,x='tip',y='total_bill',hue='sex')
plt.show()
```

```python
#Draw the scatterplot based on gender and day and size of group from tips data.
sns.scatterplot(data=df,x='tip',y='total_bill',hue=
                'sex',style='day',size='size')
plt.show()
#Hue :- assigns different colors based on category --> always provide categorical column in hue
```

```python
# Line by using the seaborn
# syntax:sns.lineplot(data=,x=,y=)
# draw the lineplot to the day and total bill
# Note: take categorical columns o x-axis
sns.lineplot(data=df,x='day',y='total_bill')
plt.show()
```

```python
# Histogram by using the Seaborn 
# Draw a histogram based on total bill and gender column
sns.histplot(data=df,x='total_bill',hue='sex')
plt.show()
```

```python
plt.hist(data=df,x='total_bill',label=df['sex'])
plt.show()
```

```python
# Boxplot:is used to detect the outliers in the data
# Box plot used to check the distribution of continuous data w.r.t. categorical data
# syntax: sns.boxplot(data=,x=,y=,hue)
sns.boxplot(data=df,x='sex',y='total_bill',hue='day')
plt.show()

# what are all component do we have in the box
# 1. 5% minimum  --> lower limit
# 2. Q1(quantile) --> 25% percentile of data
# 3. Q2           --> 50% PERCENTILE of the data
# 4. Q3           --> 75% percentile of the data
# 5. 95% maximum  --> 95% --> upper limit
```

```python
# violinplot
sns.violinplot(data=df,x='sex',y='total_bill',hue='day')
plt.show()
```

```python
# Distributional representation
# Displot: the displot support to visualize the distribution of your address
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
# Find the distribution of total_bill according to time wise from the tips data. 
sns.displot(data=df,x='total_bill',col='time',kde=True)
plt.show()

# kde--> is kernal density estimation curve that define distribution of dataset
```

```python
# Emperical cumalative distribution function gaves the step by wise increasing and decreasing distribution curve.
sns.displot(data=df,x='total_bill',col='time',kind='ecdf',hue='smoker')
plt.show()
```

```python
# Distribution of categorical data
# Draw a scatterplot for the total_blil and the day 
sns.scatterplot(data=df,x='day',y='total_bill')
plt.show()
```

```python
sns.catplot(data=df,x='day',y='total_bill',kind='swarm',hue='smoker')
plt.show()
```

```python
# statistical estimation plot
# Statistical estimation plot used to check linear relationship in the datasets
sns.lmplot(data=df,x='total_bill',y='tip',hue='smoker')
plt.show()
```
