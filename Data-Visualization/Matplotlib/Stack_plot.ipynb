```python
# Stack_plot:
# A stack plot is used to show how multiple data series contribute to a total over time.
# It displays the cumulative values with areas stacked on top of each other.
```

```python
# Time spends on different activities over days
import matplotlib.pyplot as plt

days=['Mon','Tues','Wed','Thus','Fri','Sat','Sun']
Studying=[6,7,4,10,5,3,9]
food_making=[2,3,1,2,1,1,2]
Watching=[3,4,3,2,3,5,2]
Sleeping=[8,7,9,8,6,7,8]

labels=['Studying','food_making','Watching','Sleeping']
colors=['blue','green','red','skyblue']

plt.figure(figsize=(10,6))
plt.stackplot(days,Studying,food_making,Watching,Sleeping,
             labels=labels,colors=colors,alpha=0.8)
plt.legend(loc='upper right')
plt.title("Weekly Activity Tracker")
plt.xlabel('Days')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2022, 2023, 2024, 2025,2026]
mobile = [100, 120, 150, 180,210]
laptop = [150, 180, 200, 220,310]
tablet = [50, 80, 100, 120,210]

plt.stackplot(
    years,
    mobile,
    laptop,
    tablet,
    labels=["Mobile", "Laptop", "Tablet"]
)

plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Sales Contribution by Product")
plt.legend(loc="upper left")

plt.show()
```
