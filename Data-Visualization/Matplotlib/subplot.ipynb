```python
# Subplot: Subplot allow you to show multiple plots in single figure side by side or in grid layout.
# this is helpful when comapring different datsets or aspects of the same data.
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[i*2 for i in x]
z=[i**2 for i in x]

plt.figure(figsize=(6,4))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("multiply of x")

plt.subplot(1,2,2)
plt.plot(x,z)
plt.title("Square of x")
plt.tight_layout() #line space
plt.show()
```
