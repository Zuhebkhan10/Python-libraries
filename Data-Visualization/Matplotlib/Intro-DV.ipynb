```python
import numpy as np

x=np.array([1,2,3,4,5,6])
y=np.array([2,4,3,9,7,10])

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"*") #--> p,,P,*,o,h,H
plt.show()
```

```python
# format the line plot 
# there are 4 types of format lines
# -  -->solid
# --  -->dashed lines
# : --->dotted line
# -. -->dash dot line

import numpy as np
x=np.array([1,2,4,5,6,8])
y=np.array([2,4,5,3,6,7])

plt.plot(x,y,"o-r")
plt.show()
```

```python
# Increase the size of the scatter and mark a different color to the scatter 
plt.plot(x,y,"o-r",ms=10,mfc="#fa9405",mec="#ea05fa")
plt.xlabel("Marks")
plt.ylabel("Entry number")
plt.title("Marks vs Entry number ")
plt.show()

#syntax:- plt.plot(x,y,"o -- y",ms=,mec=,mfc=)
```

```python
# Scatter plot
plt.scatter(x,y,color='#ea05fa',s=60)
plt.xlabel("Marks")
plt.ylabel("Entry number")
plt.title("Marks vs Entry number ")
plt.show()
```

```python
#Histogram --it computes based on frequency
data=[20,50,80,25,120,140,150]
plt.hist(data)
plt.show()
# There is no "y" axis in hist plot because it computes based on frequrncy
#  Syntax plt.hist(data,color='#ea05fa',bins=1)
```
