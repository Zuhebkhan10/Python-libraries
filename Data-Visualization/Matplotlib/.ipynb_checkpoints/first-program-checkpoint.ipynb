```python
import matplotlib.pyplot as plt

# Data
x = [1,2,3,4,5]
y = [10,20,25,30,40]

# Create a line plot
plt.plot(x,y)

# Add title and labels
plt.title("My First Matplotlib Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.style.use('ggplot')

# Show the graph
plt.show()
```

```python
# ploting charts using matplotlib
import matplotlib.pyplot as plt

years=[2016,2017,2018,2019,2020,2021,2022,2023,2024,2025,2026]
runs=[1020,780,670,610,910,1020,1130,855,1090,1300,1250]

plt.plot(years,runs)
plt.xlabel("Years")
plt.ylabel("Runs Scores")
plt.title("Virat Kohli yearly runs")
plt.style.use("classic")
plt.show()

```

```python
# Line width and layout tweeks
years=[2016,2017,2018,2019,2020,2021,2022,2023,2024,2025,2026]
Kohli=[973,780,670,610,910,1020,1130,855,1090,1300,1250]
Rohit=[550,600,440,701,450,750,980,790,1020,1190,1210]

plt.plot(years,Kohli,linewidth=3,label='Kohli')
plt.plot(years,Rohit,linewidth=2,label='Rohit')

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Compare the two player through line chart")
plt.legend() 
plt.grid(True)
plt.style.use("classic") # use the style format

plt.tight_layout()
plt.show()
```
