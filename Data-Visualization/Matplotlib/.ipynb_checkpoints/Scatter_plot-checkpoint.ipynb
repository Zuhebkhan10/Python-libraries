```python
# Scatter_plot: Scatterplot are used to show the relationship between two varibales.
# let's assume--> we are plotting the study hours vs exams scores for a group of students.
```

```python
import matplotlib.pyplot as plt

study_hours=[1,2,4,9,4,5,3,6,5,8,6]
exams_scores=[55,60,33,90,59,67,77,68,85,56,92]

plt.scatter(study_hours,exams_scores)
plt.title('Study hours vS Exam score')

plt.xlabel('Study hours')
plt.ylabel('Exam Scores')
plt.grid('True')
plt.show()
```

```python
# Detect overweight and underweight people using Height vs Weight
# We use a scatter plot to identify:
# Underweight zone
# Normal zone
# Overweight zone

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('height_weight.csv')

plt.scatter(df['height'],df['weight'])

plt.axhline(y=60,linestyle='--',color='r') #underweight
plt.axhline(y=60,linestyle='--') #Normalweight
plt.axhline(y=80,linestyle='--')#overweight

plt.xlabel('Height (cm)')
plt.ylabel("Weight (kg)")
plt.title("Health risk detection using height vs weight")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("height_weight.csv")

# Separate groups
male = data[data["Group"] == "Male"]
female = data[data["Group"] == "Female"]

# Scatter plots with different colors
plt.scatter(male["height"], male["weight"], label="Male")
plt.scatter(female["height"], female["weight"], label="Female")

# Labels and title
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight by Gender")

# Legend for clarity
plt.legend()

plt.grid(True)
plt.show()



# To compare two groups in a single scatter plot, we used different colors and added a legend for clear identification.
# This approach improves readability and helps in making better group-wise comparisons.
```
