```python
# add Eager and style slice with wedgeprops
import matplotlib.pyplot as plt

wicket = [10, 15, 20, 5]
labels = ["Bowled", "Caught", "LBW", "Run Out"]
colors = ["red", "blue", "green", "orange"]

plt.pie(wicket,
        labels=labels,
        colors=colors,
        wedgeprops={"edgecolor":'black','linewidth':2,"linestyle":"--"})
plt.show()
```

```python
# Interview tip (1 line)

# wedgeprops is used in a pie chart to customize the slice border style such as color, width, 
# and line pattern in Matplotlib.

# What this does

# wedgeprops styles the slice borders:
# edgecolor → black border
# linewidth → thickness of border
# linestyle → dashed border (--)


# Important Pie Chart Parameters:
    
# 1️⃣ labels
# Adds category names to each slice
# labels=["A","B","C"]

# 2️⃣ colors
# Sets slice colors
# colors=["red","blue","green"]

# 3️⃣ autopct
# Shows percentage on slices
# autopct="%1.1f%%"

# 4️⃣ explode
# Separates a slice from the pie
# explode=[0, 0.1, 0, 0]

# 5️⃣ startangle
# Rotates the pie chart
# startangle=90

# 6️⃣ shadow
# Adds shadow effect
# shadow=True

# 7️⃣ wedgeprops
# Styles slice borders (edge color, width, line style)
# wedgeprops={"edgecolor":"black","linewidth":2}

# pctdistance
# Moves percentage text inside/outside
# pctdistance=0.8

# labeldistance
# Moves labels closer or farther
# labeldistance=1.1
```

```python
import matplotlib.pyplot as plt

brands = ["Samsung", "Apple", "Others"]
market_share = [35, 30, 35]
colors = ["blue", "grey", "orange"]

plt.pie(
    market_share,
    labels=brands,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"edgecolor": "black"}
)

plt.title("Smartphone Market Share")
plt.show()


# Interview-ready conclusion (say this):

# The pie chart shows that Samsung and the Others category are leading with equal market share, while
# Apple is slightly behind. This indicates a competitive market with growth opportunities, especially 
# in the fragmented ‘Others’ segment.
```
