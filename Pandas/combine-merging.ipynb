```python
# Combining or merging the dataframes
import pandas as pd 

d1=pd.DataFrame({
    "keys":['a','b','c','c','d','e'],
    "num1":range(6)
})
d2=pd.DataFrame({
    "keys":['b','c','e','F'],
    "num2":range(4)
})
print(d1)
print(d2)

```

```python
#merge d1 and d2
pd.merge(d1,d2) #Inner join :Gets common data from both the tables


```

```python
#Left join: get all the information from the first table and matched information from 2nd table.
pd.merge(d1,d2,how="left")
```

```python
# Right join: get all the information from the firsttable and matched informatio from 1st table
pd.merge(d1,d2,how="right")
```

```python
#get all the information from both tables.
pd.merge(d1,d2,how='outer')
```
