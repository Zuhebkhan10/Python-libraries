```python
# Pivot table
# pivot table in pandas is a tool to summarzie and analysze tge data
import pandas as pd

data={
    "students":["a","b","c","d","a","b","c"],
    "subjects":["Pyhton","Sql","PowerBI","Excel","Java","ML","AI"],
    "marks":[80,34,67,68,90,77,55],
    "sem":["sem-1","sem-1","sem-1","sem-2","sem-2","sem-2","sem-2"]
}

#covert above data into df
df=pd.DataFrame(data)
df

```

```python
pd.pivot_table(df,values="marks",index="students",columns="subjects",aggfunc="max")
```

```python
# Group by
# group bu function is applicable with aggregate function

# print the min marks of student
df.groupby("students")['marks'].min()
```

```python
# Applying function dataframe
# apply a condition on df those who recieve marks greater than 85 they are eligible
def  eligibility(num):
    if num>=75:
        return "Eligible"
    else:
        return "Not Eligible"

df['marks'].apply(eligibility)
```

```python
#applying the function is dataframe
df['eligiblity']=df['marks'].apply(eligibility)
df
```
