```python
# Handling the missing data
from numpy import nan as NA
import pandas as pd
df=pd.DataFrame([[1,2,4],[20,NA,30],[NA,NA,NA]])
print(df)
```

```python
#check the null values.
df.isnull()
```

```python
#how can drop the null values.
df.dropna()
```

```python
# delete the row which have entire null values.
df.dropna(how="all")
```

```python
# replace the 2 with null values
df[1]=90
df
```

```python
#delete the columns which is having all the null values.
df.dropna(how="all",axis=1)
```

```python
import pandas as pd
from numpy import nan as NA
df=pd.DataFrame([[15,25,40],[10,NA,20],[NA,NA,NA]])
print(df)  
#delete the null value row if the count of non-null values is less than 3. 
# df.dropna(axis=0,thresh=2,inplace=True)

# df.dropna (how="any/all",axis=0/1,thresh=count of null values,inplace=True )
```

```python
# fillna(): fills missing values with a specific value or method.
#df.fillna(value)
df.fillna(0)
```

```python
#df.fillna(value,method-'ffill',limit=count of fill null values,inplace=True)
df.fillna(method='ffill',limit=1)

# ffill(): fills missing values using the previous valid value.
```

```python
# fill the null value with mean
df.fillna(df.mean())
```
